## Forces on middle line

```python
sides_angle = np.deg2rad(53)
# Acting forces (Reactions * -1) from ATENA
F_m_y = 111 # [kN]
F_m_z = -28 # [kN]
F_m_yl = F_m_y * np.cos(sides_angle) + F_m_z * np.sin(sides_angle) # Will push the L-profile up if positive
F_m_zl = F_m_y * np.sin(sides_angle) - F_m_z * np.cos(sides_angle) # Will go directly to the supporting steel body
F_m_yl, F_m_zl
```

## Forces on inclined lines

```python
# Acting forces (Reactions * -1) from ATENA
F_l_y = 30.8 # [kN]
F_l_z = -15.6 # [kN]

# Transforming to steel support surface coords yl, zl
F_l_yl = F_l_y * np.cos(sides_angle) + F_l_z * np.sin(sides_angle) # Will push the L-profile up if positive
F_l_zl = F_l_y * np.sin(sides_angle) - F_l_z * np.cos(sides_angle) # Will go directly to the supporting steel body
F_l_yl, F_l_zl
```

```python
# This is the force that will push the L-profile of the side edges up (with direction perpendicular to line as seen)
F_l_yl_prime = F_l_yl * np.cos(np.deg2rad(37))
F_l_yl_prime
```

## 1 bolt $\phi10$ can take 10 kN shear force

```python
# Steel S 235
f_y = 235
tau_y = f_y / np.sqrt(3)
A = np.pi * (10/2)**2
F = tau_y * A / 1000
F
```

=> For inclined parts one bolt 10 would be even enough

```python
# Steel S 235
f_y = 235
tau_y = f_y / np.sqrt(3)
A = np.pi * (16/2)**2
F = tau_y * A / 1000
2*F
```
