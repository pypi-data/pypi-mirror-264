```python
import cv2
import pytesseract
import os

# set the path to the folder containing the test photos
path = "D7_photos/"

# initialize an empty list to store the force values
force_list = []

# read the first image file to allow user to select the region of interest
filename = os.listdir(path)[0]
print(filename)
if filename.lower().endswith(".jpg") or filename.lower().endswith(".png"):
    print('true')
    # read the image file
    image = cv2.imread(path + filename)
    # create a window to display the image and allow user to select the region of interest
#     cv2.namedWindow("image")
#     roi = cv2.selectROI("image", image, fromCenter=False, showCrosshair=True)
#     cv2.destroyWindow("image")
    roi = (3200, 3800, 1800, 700) # x, y, w, h
    # extract the region of interest containing the force value
    force_roi = image[int(roi[1]):int(roi[1]+roi[3]), int(roi[0]):int(roi[0]+roi[2])]
    
    # convert the region of interest to grayscale
    force_roi = cv2.cvtColor(force_roi, cv2.COLOR_BGR2GRAY)
    
    # apply a threshold to the grayscale image to make the text more readable
    force_roi = cv2.threshold(force_roi, 50, 100, cv2.THRESH_BINARY)[1]
    
#     cv2.imshow('image',force_roi)
#     cv2.waitKey(0)
    
    # use pytesseract to extract the text from the thresholded image
    force_text = pytesseract.image_to_string(force_roi, config='--psm 10 --oem 3 -c tessedit_char_whitelist=0123456789.')
    
    # convert the force value to float and append it to a list
    force_list.append(force_text)

# loop through all the files in the folder
for filename2 in os.listdir(path):
    print(filename2)
    # check if the file is an image file
    if filename2.lower().endswith(".jpg") or filename2.lower().endswith(".png"):
        # read the image file
        print(path + filename2)
        image = cv2.imread(path + filename2)
        # use the same region of interest for all following images
        force_roi2 = image[int(roi[1]):int(roi[1]+roi[3]), int(roi[0]):int(roi[0]+roi[2])]
        # convert the region of interest to grayscale
        force_roi2 = cv2.cvtColor(force_roi2, cv2.COLOR_BGR2GRAY)
        # apply a threshold to the grayscale image to make the text more readable
        force_roi2 = cv2.threshold(force_roi2, 40, 100, cv2.THRESH_BINARY)[1]
        
        cv2.imshow('image',force_roi2)
        cv2.waitKey(0)
        
        # use pytesseract to extract the text from the thresholded image
        force_text = pytesseract.image_to_string(force_roi2, config='--psm 10 --oem 3 -c tessedit_char_whitelist=0123456789.')
        # convert the force value to float and append it to a list
        force_list.append(force_text)

# print the list of force values in the order of the photo files
print(force_list)
```
