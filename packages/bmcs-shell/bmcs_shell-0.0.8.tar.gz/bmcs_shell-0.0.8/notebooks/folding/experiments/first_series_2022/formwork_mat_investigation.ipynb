```python
%matplotlib widget
import numpy as np
import k3d
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
```

```python
def print_p_and_q(E, t=2, u=2):
    # u as tolerance in [mm] 
    # t thickness in [mm] 
    l = 600 # [mm]
    b = 200 # [mm]
    I = b * t**3/12
    u = 2 # [mm] tolerance
    q = u/((5/384)*l**4/(E*I))
    p = u/((1/48)*l**3/(E*I))
    print('q=' + str(round(q*1000 * 0.1, 3)) + ' kg/m' + ', p=' + str(round(p * 0.1, 2)) + ' kg') 
```

```python
# Stahl 2 mm (cheaper than alu)
E = 200000 # [MPa]
print_p_and_q(E, t=2)
```

```python
# Alu 2 mm (AlMg3)
E = 70000 # [MPa], see AlMg3 https://www.alu-sale.de/werkstoffeigenschaften
print_p_and_q(E, t=2)
```

```python
# Lexan 5 mm (AlMg3)
E = 2300 # [MPa], see https://expresszuschnitt.de/lexan-polycarbonat-zuschnitt
print_p_and_q(E, t=5)
```

```python
# PVC (AlMg3)
E = 600 # [MPa], see https://expresszuschnitt.de/lexan-polycarbonat-zuschnitt
print_p_and_q(E, t=5)
```
