```python

df=pd.DataFrame({'id':['a','b','c','d','e','','f','f'],
                  'balance':[10,20,0,21,15,10,20,25],
                  'country':['sg','cn','ca','np','in','in','np','in']})

df
```

```python
df[['id','balance']].agg_df()
```

```python
df[['id','balance']].agg_df(type='sum',count=False)
```

```python
df[['id','balance']].agg_df(type='mean',count=True)
```

```python
df[['id','balance']].agg_df(type='min',count=True)
```

```python
df[['id','balance']].agg_df(type='mean',count=False)
```

```python
df[['id','balance']].agg_df(count=True)
```

```python
help(pd.DataFrame.agg_df)
```

```python
df[['id','balance']].agg_df(type='sum',update_col=True,count=True)
```
