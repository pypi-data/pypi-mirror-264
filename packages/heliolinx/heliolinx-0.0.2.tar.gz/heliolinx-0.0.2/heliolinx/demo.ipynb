```python
df = pd.read_csv('../tests/testin_mt01.csv')
sa = np.zeros(len(df), dtype=solardg.hldet)
for l, r in [
             ("idstring", "ObjID"), ("MJD", "FieldMJD"), ("RA", "AstRA(deg)"), ("Dec", "AstDec(deg)"),
             ("mag", "MaginFilterTrue"), ("band", "filter"), ("obscode", "obscode")
            ]:
    sa[l] = df[r]
sa.sort(order=['MJD','obscode','RA'])
pd.DataFrame(sa)
```

```python
config = heliohypy.MakeTrackletsConfig()
config.imagerad = 5.0
config.minarc = 1.0

(dets, tracklets, trac2det) = heliohypy.makeTracklets(config,sa,newimage)
```

```python
def df_to_numpy(df, dtype):
    sa = np.zeros(len(df), dtype=dtype)
    for col in df.columns:
        sa[col] = df[col]
    return sa

def det_to_numpy(df):
    return df_to_numpy(df, dtype=solardg.hldet)

def vis_to_numpy(df):
    return df_to_numpy(df, dtype=solardg.hlimage)

sa2 = det_to_numpy(sa_df)
newimage2 = vis_to_numpy(newimage_df)

(dets, tracklets, trac2det) = heliohypy.makeTracklets(config,sa2,newimage2)
```
